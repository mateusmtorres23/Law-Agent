```python
from agno.agent import Agent
from agno.models.ollama import Ollama
from IPython.display import Markdown, display
import pymupdf4llm
import os

FILE_PATH = "assets/"

def read_pdf(file_name: str) -> str:
    """
    Read a PDF file and return its content in markdown for analysis.
    use his tool to gain insights from legal documents, contracts, or case studies the user sent.
    """

    caminho_completo = os.path.join(FILE_PATH, file_name)

    if not os.path.exists(caminho_completo):
        return f"Erro: O arquivo {file_name} não foi encontrado na pasta {FILE_PATH}."
    
    return pymupdf4llm.to_markdown(caminho_completo)

law_agent = Agent(
    model=Ollama(id="qwen3.5:2b", options={
            "num_gpu": 99,        
            "num_ctx": 4096,      
            "num_thread": 8,
            "low_vram": False    
        }),
    tools=[read_pdf],
    description="""A legal assistant that can answer questions about the law and provide legal advice.""",
    markdown=True
)

response = law_agent.run("O que a autora (motorista) alega sobre a supervisora no arquivo 'peticao_trabalhista.pdf'?")
display(Markdown(response.content))
```
