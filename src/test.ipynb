```python
from agno.agent import Agent
from agno.models.ollama import Ollama
from IPython.display import Markdown, display

law_agent = Agent(
    model=Ollama(id="qwen3.5:2b", options={
            "num_gpu": 99,        
            "num_ctx": 1024,      
            "num_thread": 8,       
            "low_vram": False    
        }),
    description="""A legal assistant that can answer questions about the law and provide legal advice.""",
    markdown=True
)

response = law_agent.run("What is the difference between a contract and an agreement?")
display(Markdown(response.content))
```
